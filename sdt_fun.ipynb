### Compute P(repeat) for each individual
We quantified individual choice history bias in terms of the probability of repeating a choice, termed P(repeat), regardless of the category of the (previous or current) stimulus. This yielded a measure of bias that ranged between 0 (maximum alternation bias) and 1 (maximum repetition bias), whereby 0.5 indicated no bias.

You have read in the introduction that an event is a well-defined subset of the sample space. Events can be classified into two categories: dependent or independent.

Independent events are events that don't impact the probability of the other event(s). Two events A and B are independent if knowing whether event A occurred gives no information about whether event B occurred.

```python
# Create function that returns probability percent rounded to one decimal place
def event_probability(event_outcomes, sample_space):
    probability = (event_outcomes / sample_space) * 100
    return round(probability, 1)
```

```python
Z = norm.ppf     
# This function turns 'proportions' into z-scores:
# Z(.5) = 0.0
# Z(.75) = 0.674
# Z(.1581) = -1.0

def sdt(hit_rate, fa_rate):
    """
    This function takes a hit rate and a false alarm rate, and computes d_prime and c
    """
    out = dict()
    out['d_prime'] = Z(hit_rate) - Z(fa_rate)
    out['criterion'] = -0.5*(Z(hit_rate) + Z(fa_rate)) #bias 
    return out
```

```python
def get_hit_fa(df):
    
    """
    gets a dataframe with subjects and stimulus columns (1 for stim present, 0 for stim absent)
    calculates hit rate and false alarm rate for each subject
    """
    subjects = []
    age = []
    fa = []
    hit = []
    miss = []
    cr = []
    d_prime = []
    c = []
    SubjectList = np.unique(df['subj_idx'])
    
    for sub in SubjectList:

        sub_df = df[df['subj_idx'] == sub]
        resp = np.unique(sub_df["response"][(sub_df["stim"] == 1) & (sub_df["accuracy"] == 1)]) 
        print(resp[0]) # get button for "yes I've seen this" per subject, button assignment changes per subject
        # create column with 1 for "yes I have seen this" (aka old) and 0 for "no have not seen this before" (aka new)
        sub_df['resp']=np.where(sub_df["response"] == resp[0],1,0)
        # & (sub_df['resp'] == 1) do I need this??? or is accuracy and target present enough?
        # 1 for hit 0 otherwise
        sub_df['hit'] = np.where((sub_df['stim']==1) & (sub_df['accuracy'] == 1), 1, 0)
        # 1 for miss and 0 otherwise
        sub_df['miss'] = np.where((sub_df['stim']==1) & (sub_df['accuracy'] == 0), 1, 0) # is miss not basically the 0s of the hit column???
        # 1 for FA 0 otherwise
        sub_df['FA'] = np.where((sub_df['stim']==0) & (sub_df['accuracy'] == 0), 1, 0)
        # 1 for cr and 0 otherwise
        sub_df['cr'] = np.where((sub_df['stim']==0) & (sub_df['accuracy'] == 1), 1, 0)
        #print(sub_df.head())
        hit_rate = sub_df['hit'].value_counts(normalize=True).to_dict()
        false_alarm = sub_df['FA'].value_counts(normalize=True).to_dict()
        miss_rate = sub_df['miss'].value_counts(normalize=True).to_dict()
        cr_rate = sub_df['cr'].value_counts(normalize=True).to_dict()
        if len(hit_rate) != 2:
            print("Hit rate", hit_rate) #don't think I need this anymore, as subjects have been eliminated
            hit_rate = 0.00000000000000001
        else:
            print("Subject", sub, "Hitrate:",np.round(hit_rate[1]*100,2), "Miss rate:", np.round(miss_rate[1]*100,2))
            print("FA rate:", np.round(false_alarm[1]*100,2), "CR rate:", np.round(cr_rate[1]*100,2))
            # all 4 rates add up to 100!
            hit_rate = hit_rate[1]
            fa_rate = false_alarm[1]
            
        dc_dict = sdt(hit_rate, fa_rate)
        
        print(dc_dict)
        #print(dc_dict['d_prime'])
        subjects.append(sub)
        age.append(np.unique(sub_df['age'])[0])
        hit.append(hit_rate)
        fa.append(fa_rate)
        miss.append(miss_rate[1])
        cr.append(cr_rate[1])
        d_prime.append(dc_dict['d_prime'])
        c.append(dc_dict['criterion'])
    
    dict_sdt = {'subject':subjects,'group':age,'hit': hit, 'miss':miss, 'cr':cr, 'FA': fa, 'd_prime':d_prime , 'criterion':c}
    
    df_sdt = pd.DataFrame(dict_sdt)
    return df_sdt
```

```python
def draw_gauss(m,  s, ax, c, label='', rnge=[-5,5]):
    """
    draws a gaussian at x, with sd s in the specified axis using color c
    """
    x = np.linspace(rnge[0], rnge[1], 1000)
    ax.plot(x, norm.pdf(x, loc=m, scale=s), c=c, label=label)
```

```python
# we define some parameters for our example plot using the above function
mean_diff = 1.2286973955287714
criterion = 0.7907228625452472
s = sqrt(mean_diff)
```

```python
# First of course make a figure
f, ax = plt.subplots(1, 1, figsize=(16,6))

# and then draw normal distributions
draw_gauss(m=-mean_diff/2, s=s, ax=ax, c='r', label='noise')
draw_gauss(m=mean_diff/2, s=s, ax=ax, c='b', label='signal')

# and, we can also plot 'c', the criterion value in the plot
ax.axvline(x=criterion,  color='k', linestyle='--', label='criterion') 
plt.legend();
```
