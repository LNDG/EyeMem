# DeepGaze II and ICF

This notebook demonstrates how to load and use the DeepGaze II and ICF models.
It needs
to be run using python3 and an up to date tensor flow installation (I use
tensorflow 1.1, but tensorflow 1.0 might work, too). Using a GPU will
considerably speed up computation.

DeepGaze expects it's input to have a resolution of 35 pixel per degree of visual angle. If your stimuli have a different resolution, you have to rescale them before processing them with DeepGaze, otherwise you will get wrong predictions.

```python
%matplotlib inline
```

```python
# importing all required packages

import numpy as np
from scipy.ndimage import zoom
from scipy.special import logsumexp
from scipy.io import savemat


import matplotlib.pyplot as plt

import pandas as pd

import seaborn as sns
sns.set_style('white')

from PIL import Image
import os

path = "../../stimuli_640x480/"
```

```python
# 35/49.15 = 0.7121
resize_ratio = 0.7121 

""" 
Resize the images by a scaling factor of 0.7121, 
from 640x480 to 455x341 pixels. Save the rescaled images
in same directory with _35pxl extension in their name

"""

def resize_aspect_fit():
    
    for subdir, dirs, files in os.walk(path):
        if subdir.endswith("/landscapes"):
            continue
        for file in files:
            if file == '.DS_Store' or file.endswith("_35pxl.png"):
                continue
            if file.endswith(".bmp") or file.endswith(".png"):
               
                image = Image.open(os.path.join(subdir, file)) 
                file_path, extension = os.path.splitext(subdir + "/" + file)
              
                new_image_height = int(image.size[0] / (1/resize_ratio))
                new_image_length = int(image.size[1] / (1/resize_ratio))

                image = image.resize((new_image_height, new_image_length), Image.ANTIALIAS)
                #image.save(file_path + "_35pxl" + extension, 'png', quality=90)


#resize_aspect_fit()
```

```python
# Importing tf as an older version, necessary to get the tf version 1.1 code working
import tensorflow.compat.v1 as tf
#tf.disable_v2_behavior() 
tf.disable_eager_execution()
```

```python
"""
Retrieve the rescaled images.
Save them in a list for further computation.

"""

import re

def retrieve_img():
    images = []
    shown_imgages = np.loadtxt('../#_of_pics.dat',delimiter=',')
    for subdir, dirs, files in os.walk(path):
        for file in files:
            if "_35pxl" in file:
                number = re.findall(r'\d+', file)
                file_number = int(number[0])
                if file_number in shown_imgages:
                   
                    images.append(os.path.join(subdir, file))
    return images
 
images = retrieve_img()
```

```python
# quick check if it worked, should contain all 300 images and each index of list being path to one image
print(len(images))
print(images[0])
lst = ['landscapes', 'streets1', 'fractals', 'naturals1', 'streets2']
image_number = []
image_cat = []
for i in range(0,len(images)):
    #print(images[i])
    image_number.append(re.findall(r"(\d+)_", images[i]))
    words_re = re.compile("|".join(lst))
    image_cat.append(re.findall(words_re, images[i]))

#print(image_number)
```

```python
# retrieving one images for further processing of image center bias
img_centerB = Image.open(images[0])
img_centerB = np.array(img_centerB)
```

DeepGaze II and ICF incorporate a center bias as a prior over the fixation distribution. Here we are going to use the center bias from the MIT1003 dataset (see notebook `Centerbias.ipynb`). You should use the best centerbias you can come up with for your dataset.

```python
# load precomputed log density 
centerbias_template = np.load('centerbias.npy')  
```

```python
# rescale to match image size
centerbias = zoom(centerbias_template, (img_centerB.shape[0]/1024, img_centerB.shape[1]/1024), order=0, mode='nearest')
# renormalize log density
centerbias -= logsumexp(centerbias)
centerbias
```

```python
import pysaliency
import pysaliency.plotting
pysaliency.plotting.visualize_distribution(centerbias)
```

The model expects all input as 4d tensors of shape `BHWC` (i.e. batch-height-
width-channel). It takes two inputs:
A batch of images and a batch of centerbias log densities.

```python
"""
Change each image from PIL image object to numpy array.
Note: After you make your changes to the array, 
you should be able to do either pic.putdata(pix) 
create a new image with Image.fromarray(pix)

"""
image_data = []
for i in range(0, len(images)):
    img = Image.open(images[i])
    img = np.array(img)
    img_data = img[np.newaxis, :, :, :]  # BHWC, three channels (RGB)
    image_data.append(img)
```

```python
centerbias_data = centerbias[np.newaxis, :, :, np.newaxis]  # BHWC, 1 channel (log density)
```

```python
# check point: one item in our dataset should have the shape (341, 455, 3) with the first two entries being the pixels
image_data[0].shape
```

Now we import the deep gaze model from the tensorflow meta-graph file

```python
tf.reset_default_graph()
```

```python
check_point = 'DeepGazeII.ckpt'  # DeepGaze II
#print(check_point)
#check_point = 'ICF.ckpt'  # ICF
new_saver = tf.train.import_meta_graph('{}.meta'.format(check_point))

input_tensor = tf.get_collection('input_tensor')[0]
centerbias_tensor = tf.get_collection('centerbias_tensor')[0]
log_density = tf.get_collection('log_density')[0]
log_density_wo_centerbias = tf.get_collection('log_density_wo_centerbias')[0]
```

```python
test = tf.train.NewCheckpointReader(check_point)
#test.get_variable_to_shape_map()
```

And finally we create a tensorflow session, restore the model parameters from
the checkpoint and compute the log density prediction for out input data:

```python
with tf.Session() as sess:
    
    new_saver.restore(sess, check_point)
    
    log_density_prediction = sess.run(log_density, {
        input_tensor: image_data,
        centerbias_tensor: centerbias_data,
    })
```

The log density predictions again are of shape `BHWC`. Since the log-densities
are just 2d, `C=1`. And since we processed only one image, `B=1`:

```python
# log density prediction should be of shape (150, 341, 455, 1) - first entry representing # of images
print(log_density_prediction.shape)
```
