```python
def t_test(df1, df2, col):
    '''returns t and corresponding p value for given df columns'''
    t,p = stats.ttest_ind(df1[col],df2[col])
    print("The difference between means", (np.mean(df1[col])-np.mean(df2[col])))
    print(col+ " t-val:", t, "p-val:", p)
    return p
```

```python
em1_test_RT_hit_p = t_test(em1_test_RT_hit_ya, em1_test_RT_hit_oa,'mean_RT_hits') 
em1_test_RT_fa_p = t_test(em1_test_RT_hit_ya, em1_test_RT_hit_oa,'mean_RT_fas') 
```

```python
def fancy_t_test(df1, df2, val):
    summary, results = rp.ttest(group1= df1[val], group1_name= "YA",
                            group2= df2[val], group2_name= "OA")
    print(summary)
    print(results)
```

```python
def plot_rt(df, val1, val2, p1, p2, title):
    
    fig=plt.figure(figsize=(10,5))
    ax1 = plt.subplot(121)
    ax2 = plt.subplot(122)

    sns.barplot(x="group", y=val1, data=df, capsize=.1, ci="sd", ax=ax1)
    sns.swarmplot(x="group", y=val1, data=df, color="0", alpha=.35, ax=ax1)

    sns.barplot(x="group", y=val2, data=df, capsize=.1, ci="sd", ax=ax2)
    sns.swarmplot(x="group", y=val2, data=df, color="0", alpha=.35, ax=ax2)

    x1, x2 = 0, 1   
    y1, h1, col1 = df[val1].max()+0.1,0.1, 'k'
    y2, h2, col2 = df[val2].max()+0.1,0.1, 'k'
    ax1.plot([x1, x1, x2, x2], [y1, y1+h1, y1+h1, y1], lw=1, c=col1)
    ax1.text((x1+x2)*.5, y1+h1, p1, ha='center', va='bottom', color=col1)
    ax2.plot([x1, x1, x2, x2], [y2, y2+h2, y2+h2, y2], lw=1, c=col2)
    ax2.text((x1+x2)*.5, y2+h2, p2, ha='center', va='bottom', color=col2)
    for p in ax1.patches:
            h1, w1, x1 = p.get_height(), p.get_width(), p.get_x()
            xy1 = (x1 + w1 / 2., h1 / 2)
            text = f'Mean:\n{h1:0.2f}'
            ax1.annotate(text=text, xy=xy1, ha='right', va='center_baseline')
    for p in ax2.patches:
            h2, w2, x2 = p.get_height(), p.get_width(), p.get_x()
            xy2 = (x2 + w2 / 2., h2 / 2)
            text = f'Mean:\n{h2:0.2f}'
            ax2.annotate(text=text, xy=xy2, ha='right', va='center_baseline')
    fig.suptitle(title)
    #plt.savefig('/Users/terlau/LNDG/EyeMem/EyeMem_1/hits_fas_RTs.png')
    ax1.set_xlabel('')
    ax1.get_shared_x_axes().join(ax1, ax2)
    ax1.set_xticklabels([])
    plt.tight_layout()
    plt.show()
```

```python
def plot_bar(df, val, p_val):
    
    fig, ax = plt.subplots(figsize=(8,6))

    sns.set(style="whitegrid")


    sns.barplot(x="group", y=val, data=df,  capsize=.1, color='lightblue') # ci="sd"
    sns.swarmplot(x="group", y=val, data=df, color="0", alpha=.35)
    # show the mean
    for p in ax.patches:
        h, w, x = p.get_height(), p.get_width(), p.get_x()
        xy = (x + w / 2., h / 2)
        text = f'Mean:\n{h:0.2f}'
        ax.annotate(text=text, xy=xy, ha='right', va='center_baseline')

    # statistical annotation
    x1, x2 = 0, 1   
    y, h, col = df[val].max()+0.09,0.09, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1, c=col)
    plt.text((x1+x2)*.5, y+h, p_val, ha='center', va='bottom', color=col)
    #plt.title("{} OA vs YA {}".format(param, model_type))
    #plt.savefig('/Users/terlau/LNDG/EyeMem/HDDM/plots/{}_test_eyemem1_{}.png'.format(model_type, param))
    plt.show()
    plt.close()
```

```python
def one_sample_t_test(df, val):
    #perform one sample t-test
    t,p = stats.ttest_1samp(df[val], 0) #alternative='greater'
    print(p)
    return p
```

```python
def corr(x,y, label1, label2):
    corr, p = pearsonr(x, y)
    print('Pearsons correlation: %.3f' % corr, "P_val:", p)
    c_s, p_s = spearmanr(x, y)
    print('Spearmans correlation: %.3f' % c_s, "P_val:", p_s)
    c_k, p_k = stats.kendalltau(x, y)
    print('Kendalls correlation: %.3f' % c_k, "P_val:", p_k)
    fig = plt.figure(figsize=(10, 7))
    sns.scatterplot(x, y)
    plt.ylabel("HDDM {}".format(label1))
    plt.xlabel("SDT {}".format(label2))
    plt.title('YA EyeMem1 DDM vs SDT'+'\nSpearmans correlation: %.3f' % c_s + " P_val: %.3f" % p_s, pad=15)
    #plt.savefig(title+".jpeg", bbox_inches='tight', dpi=100) # bbox_inches='tight' only works with .jpeg
    plt.show()
```

```python
def corr_2(x,y, x1, x2, y1, y2, label1, label2, a, b):
    
    c_s, p_s = spearmanr(x, y)
    print('Spearmans correlation overall subjects: %.3f' % c_s, "P_val:", p_s)
    
    c_y, p_y = spearmanr(x1,y1)
    c_o, p_o = spearmanr(x2,y2)
    # Plot
    plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':100})
    # correlation YA
    plt.scatter(x1, y1, label='YA Correlation = {} p = {}'.format(np.round(c_y,3),np.round(p_y,3)))
    # correlateion OA 
    plt.scatter(x2, y2, label='OA Correlation = {} p = {}'.format(np.round(c_o,3),np.round(p_o,3)))


    plt.title('EyeMem1 {} vs {}'.format(a,b), pad=15)
    plt.ylabel("HDDM {}".format(label1))
    plt.xlabel("SDT {}".format(label2))
    plt.legend()
   # plt.savefig('/Users/terlau/LNDG/eyemem1_corr_{}_{}.jpeg'.format(a,b), bbox_inches='tight', dpi=100) # bbox_inches='tight' only works with .jpeg
    plt.show()
```
